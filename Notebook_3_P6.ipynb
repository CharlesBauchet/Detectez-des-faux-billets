```python
# Importations des librairies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import plotly.express as px
import scipy.stats as st
from sklearn import preprocessing, cluster, metrics, decomposition
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
# Stockage de la variable 'id' dans une variable pour pourvoir l'injecter plus tard
df_example = pd.read_csv('P6_Data/test_notes.csv')
df_example = df_example['id']
```

```python
# Création de notre modèle de régression logistique
#Importation de notre dataframe
df_cls = pd.read_csv('df_cls')

#Préparation des données
X = df_cls.copy()
X = X.iloc[:, 3:] # Mise à l'écart de la variable qualitatif
y = df_cls.iloc[:, 2] # Stockage des valeurs de la variable 'clusters' dans une nouvelle variable

# Séparation entre les données d'entrainement et nos données test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Entrainement de nos données
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_train, y_train)

# Prédisons les valeurs de notre variable de test
y_predict = lr.predict(X_test)

# Projetons notre prédiction dans une matrice de confusion
conf_matrix = confusion_matrix(y_test, y_predict)
```

```python
# Préparation des données
df_test = input('renseigner le nom du dataset :')
df = pd.read_csv(df_test)
Xi = df.copy()
Xi = Xi.iloc[:, :-1]

# Centrage et réduction de nos données, Standardisations de nos valeurs
std_scale = preprocessing.StandardScaler().fit(Xi)
X_scaled = std_scale.transform(Xi)

# Utilisation du modéle de prédiction
proba = lr.predict_proba(X_scaled)[:, 1]

# Transformation de notre variable en 'séries' pour pouvoir l'ajouter à notre dataframe
proba = pd.Series(proba.round(3), name = 'probabilite')

# Ajout de cette nouvelle variable à notre dataframe
df_proba = pd.concat([Xi, proba], axis = 1)

# Création d'une boucle 'for' pour ajouter une colonne qui va dire si ce sont de vrais ou faux billet selon la probabilité calculé
nature = []
for i in df_proba['probabilite'] >= .5:
    if i is True :
        nature.append('Vrai billet')
    else :
        nature.append('Faux billet')
    
df_proba['nature'] = nature
df_proba
```

```python
# Voici à quoi ressemble le dataframe finale
df_finale = pd.concat([df_proba, df_example], axis = 1)
df_finale
```
